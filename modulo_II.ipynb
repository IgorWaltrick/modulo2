# condicional


```python
# if
if 5 > 10 :
    print('verdade')
```

```python
# if / else
num1= 10
num2= 5
if num1 > num2 :
    print('verdade')
else:
  print('num2 é maior')
```

```python
# if / elif / else
num1= 10
num2= 5
if num1 > num2 :
    print('verdade')
elif num1 < num2 :
  print('num2 é maior')
else:
  print('os 2 são iguais')

```

```python
# if / elif / else + operadoes logicos
num1= 10
num2= 5
num3 = 7
if (num1 > num2) and (num1 > num3):
    print('num1 é o maior')

else:
  print('num1 não é o maior')

```

```python
lista = ['teste 1', 'teste 2']
```

```python
#for simples
for val in lista:
  print(val)
```
